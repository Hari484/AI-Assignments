## Assignment 4:Data Analysis and Visualization Using Pandas

```python
# Install necessary libraries
!pip install pandas matplotlib seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset from the UCI Machine Learning Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, names=columns)

# Display the first few rows of the dataframe
print("First few rows of the dataframe:")
print(df.head())

# Display basic information about the dataframe
print("\nDataframe Information:")
df.info()

# Display summary statistics of the dataframe
print("\nSummary Statistics:")
print(df.describe())

# Display the unique species in the dataset
print("\nUnique Species:")
print(df['species'].unique())

# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values)

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

# Remove duplicate rows if any
df = df.drop_duplicates()

# Display the first few rows of the cleaned dataframe
print("\nFirst few rows of the cleaned dataframe:")
print(df.head())

# Set the style for the plots
sns.set(style="whitegrid")

# Pairplot to visualize the pairwise relationships in the dataset
print("\nPairplot:")
sns.pairplot(df, hue='species')
plt.show()

# Boxplot to visualize the distribution of sepal lengths for each species
print("\nBoxplot:")
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Distribution of Sepal Lengths by Species')
plt.show()

# Heatmap to visualize the correlation matrix
print("\nHeatmap:")
plt.figure(figsize=(10, 6))
sns.heatmap(df.drop(columns=['species']).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Analysis of pairplot
print("\nPairplot Analysis:")
print("The pairplot shows distinct clusters for each species, indicating that the features (sepal length, sepal width, petal length, petal width) are good discriminators for classifying the species.")

# Analysis of boxplot
print("\nBoxplot Analysis:")
print("The boxplot shows that Setosa has the smallest median sepal length, while Virginica has the largest. The distribution of sepal lengths for Versicolor overlaps with the other two species.")

# Analysis of heatmap
print("\nHeatmap Analysis:")
print("The heatmap shows that petal length and petal width are strongly positively correlated, indicating that as the petal length increases, the petal width also increases. Sepal length and sepal width have a weak negative correlation.")
```
